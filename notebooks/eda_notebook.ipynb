```python
import pandas as pd
import numpy as np
#import os
#print("Current working directory:", os.getcwd())
#print("Files in current directory:", os.listdir('.'))
#print("Files in data/ folder:", os.listdir('data') if os.path.exists('data') else "data/ folder not found")
print("Loading datasets...")
netflix = pd.read_csv('../data/netflix_titles.csv')
tmdb    = pd.read_csv('../data/tmdb_5000_movies.csv')

print(f"Netflix original shape: {netflix.shape}")
print(f"TMDB original shape:    {tmdb.shape}\n")
```

```python
# ── 2.2 Define EMEA countries (expanded practical list) ─────────────────────
# Sources: Common business/regional definitions (e.g., Europe + Middle East + Africa)
# These names are formatted to match typical Netflix 'country' values (case-sensitive string matching)
emea_countries = [
    # Europe (major ones + some others common in Netflix originals)
    'United Kingdom', 'France', 'Germany', 'Spain', 'Italy', 'Netherlands', 'Belgium', 
    'Sweden', 'Denmark', 'Norway', 'Finland', 'Poland', 'Portugal', 'Ireland', 
    'Switzerland', 'Austria', 'Czech Republic', 'Hungary', 'Romania', 'Greece', 
    'Turkey', 'Russia', 'Ukraine', 'Belgium', 'Luxembourg', 'Iceland', 
    
    # Middle East / West Asia (key Netflix markets)
    'Israel', 'Saudi Arabia', 'United Arab Emirates', 'Egypt', 'Lebanon', 'Jordan', 
    'Qatar', 'Kuwait', 'Bahrain', 'Oman', 'Iran', 'Iraq', 
    
    # Africa (focus on countries with notable Netflix content/production)
    'South Africa', 'Nigeria', 'Kenya', 'Ghana', 'Morocco', 'Algeria', 'Tunisia', 
    'Egypt', 'Ethiopia', 'Senegal', 'Uganda', 'Zambia', 'Zimbabwe', 'Angola', 
    'Cameroon', 'Ivory Coast',  # often listed as "Côte d'Ivoire" or "Ivory Coast"
    
    # Additional common ones if your data includes them
    'Poland', 'Croatia', 'Serbia', 'Bulgaria', 'Slovakia', 'Slovenia', 
    'Bosnia and Herzegovina', 'Albania', 'North Macedonia'  # sometimes "Macedonia"
]

# Optional: Make it a set for faster lookup (recommended for filtering)
emea_countries_set = set(emea_countries)

# Print to verify
print(f"Number of EMEA countries defined: {len(emea_countries)}")
print("First 10:", emea_countries[:10])
emea_set = set(emea_countries)  # faster lookup
print(f"EMEA countries defined: {len(emea_countries)}\n")
```

```python
# ── 2.3 Filter Netflix for EMEA ─────────────────────────────────────────────
# Handle multi-country strings like "United Kingdom, France, United States"
def is_emea_country(country_str):
    if pd.isna(country_str):
        return False
    countries = [c.strip() for c in str(country_str).split(',')]
    return any(c in emea_set for c in countries)

netflix['is_emea'] = netflix['country'].apply(is_emea_country)
netflix_emea = netflix[netflix['is_emea']].copy()

print(f"Total Netflix titles:       {len(netflix):,}")
print(f"EMEA-filtered titles:       {len(netflix_emea):,} ({len(netflix_emea)/len(netflix):.1%})")
print(f"EMEA titles after 2015 only:{len(netflix_emea[netflix_emea['release_year'] >= 2015]):,}\n")

# Optional: restrict to recent years (recommended for relevance)
netflix_emea = netflix_emea[netflix_emea['release_year'] >= 2015].copy()
```

```python
# ── 2.4 Handle genres (split & explode) ─────────────────────────────────────
# 'listed_in' is comma-separated → create list + exploded version
netflix_emea['genres_list'] = netflix_emea['listed_in'].str.split(', ')
netflix_emea_exploded = netflix_emea.explode('genres_list').copy()
netflix_emea_exploded['genres_list'] = netflix_emea_exploded['genres_list'].str.strip()

print("Top 10 genres in EMEA Netflix content:")
print(netflix_emea_exploded['genres_list'].value_counts().head(10))
print()
```

```python
# ── 2.5 Separate Movies vs TV Shows ─────────────────────────────────────────
movies_emea = netflix_emea[netflix_emea['type'] == 'Movie'].copy()
tv_emea     = netflix_emea[netflix_emea['type'] == 'TV Show'].copy()

print(f"EMEA Movies: {len(movies_emea):,}")
print(f"EMEA TV Shows: {len(tv_emea):,}\n")# ── 2.6 Prepare TMDB for merge ──────────────────────────────────────────────
# Extract year, standardize title
tmdb = tmdb.copy()
tmdb['release_year'] = pd.to_datetime(tmdb['release_date'], errors='coerce').dt.year
tmdb['title_lower']  = tmdb['title'].str.lower().str.strip()

movies_emea['title_lower'] = movies_emea['title'].str.lower().str.strip()
```

```python
# ── 2.6 Prepare TMDB for merge ──────────────────────────────────────────────
# Extract year, standardize title
tmdb = tmdb.copy()
tmdb['release_year'] = pd.to_datetime(tmdb['release_date'], errors='coerce').dt.year
tmdb['title_lower']  = tmdb['title'].str.lower().str.strip()

movies_emea['title_lower'] = movies_emea['title'].str.lower().str.strip()
```

```python
# ── 2.7 Merge Netflix EMEA movies with TMDB (budget/revenue) ────────────────
merged = pd.merge(
    movies_emea,
    tmdb[['title_lower', 'release_year', 'budget', 'revenue', 
          'production_countries', 'genres', 'popularity', 'vote_average']],
    on=['title_lower', 'release_year'],
    how='inner',
    suffixes=('_netflix', '_tmdb')
)

# Keep only rows with valid budget > 0
merged = merged[merged['budget'] > 0].copy()

# Calculate ROI proxy
merged['roi'] = np.where(
    merged['budget'] > 0,
    (merged['revenue'] - merged['budget']) / merged['budget'],
    np.nan
)

print(f"Merged rows with budget data: {len(merged):,}")
print(f"Match rate (vs EMEA movies): {len(merged)/len(movies_emea):.1%}\n")

# ── 2.8 Quick inspection ─────────────────────────────────────────────────────
print("Merged columns:")
print(merged.columns.tolist())
print("\nSample merged data (first 3 rows):")
display(merged[['title', 'release_year', 'country', 'budget', 'revenue', 'roi', 'vote_average']].head(3))

print("\nBasic stats on merged finance data:")
print(merged[['budget', 'revenue', 'roi']].describe().round(2))
```

```python

# Create export folder if missing
export_dir = '../data_export'
os.makedirs(export_dir, exist_ok=True)

print("Export folder ready:", export_dir)

# ── 1. Top 10 EMEA countries ────────────────────────────────────────────────
top_countries = duck("""
    SELECT 
        country,
        COUNT(*) AS num_shows
    FROM netflix_emea
    GROUP BY country
    ORDER BY num_shows DESC
    LIMIT 10
""")

print("Top 10 EMEA countries:")
display(top_countries)
top_countries.to_csv(f'{export_dir}/top_emea_countries.csv', index=False)
print("→ Exported: top_emea_countries.csv\n")

# ── 2. Average ROI by genre (high-budget movies) ─────────────────────────────
roi_by_genre = duck("""
    WITH high_budget AS (
        SELECT 
            genres, 
            revenue, 
            budget,
            CAST(revenue - budget AS DOUBLE) / budget AS roi
        FROM merged
        WHERE budget > 10000000
          AND revenue > 0
    )
    SELECT 
        genres,
        AVG(roi)          AS avg_roi,
        COUNT(*)          AS num_movies
    FROM high_budget
    GROUP BY genres
    HAVING COUNT(*) >= 2
    ORDER BY avg_roi DESC
    LIMIT 10
""")

print("Average ROI by genre (high-budget):")
display(roi_by_genre)
roi_by_genre.to_csv(f'{export_dir}/roi_by_genre.csv', index=False)
print("→ Exported: roi_by_genre.csv\n")

# ── 3. Top-rated per country (with rank) ────────────────────────────────────
top_rated_per_country = duck("""
    SELECT 
        country,
        title,
        rating,
        release_year,
        ROW_NUMBER() OVER (PARTITION BY country ORDER BY rating DESC, release_year DESC) AS rank_in_country
    FROM netflix_emea
    WHERE rating IS NOT NULL
    LIMIT 20
""")

print("Top-rated per country (sample with rank):")
display(top_rated_per_country)
top_rated_per_country.to_csv(f'{export_dir}/top_rated_per_country.csv', index=False)
print("→ Exported: top_rated_per_country.csv\n")

# ── 4. Year-over-year growth ────────────────────────────────────────────────
yearly_growth = duck("""
    WITH yearly AS (
        SELECT release_year, COUNT(*) AS num_titles
        FROM netflix_emea
        GROUP BY release_year
    )
    SELECT 
        release_year,
        num_titles,
        LAG(num_titles) OVER (ORDER BY release_year) AS prev_year,
        num_titles - LAG(num_titles) OVER (ORDER BY release_year) AS growth,
        ROUND( 
            (num_titles * 1.0 / LAG(num_titles) OVER (ORDER BY release_year) - 1) * 100, 
            1 
        ) AS growth_pct
    FROM yearly
    ORDER BY release_year
""")

print("Year-over-year growth:")
display(yearly_growth)
yearly_growth.to_csv(f'{export_dir}/emea_yearly_growth.csv', index=False)
print("→ Exported: emea_yearly_growth.csv\n")

# Optional: Re-export main files if needed
#netflix_emea.to_csv(f'{export_dir}/netflix_emea.csv', index=False)
#merged.to_csv(f'{export_dir}/merged_finance.csv', index=False)
#print("Main files also exported/updated")
```
